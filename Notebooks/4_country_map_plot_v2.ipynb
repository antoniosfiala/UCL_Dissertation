```python
# file path
base_shape_file = f"{path_map}ne_50m_admin_0_countries/ne_50m_admin_0_countries.shp"

# geopandas dataframe
gdf_map = gpd.read_file(base_shape_file)[["ADM0_A3","ADMIN","geometry"]]
gdf_map.columns = ["iso_3","country_name","geometry"]
```

### Part 2 | BRI country list
|| [0| Default imports](#ID_top) || [1|Part1](#ID_part1) || [2|Part2](#ID_part2) || [3|Part3](#ID_part3) || [4|Part4](#ID_part4) || [5|Part5](#ID_part5) ||

```python
# BRI country list
os.listdir(f"{path_store}")[3] #store_bri_countries_Dumor_Yao_20200621_1901.csv.gzip

df_bri_countries = pd.read_csv(f"{path_store}store_bri_countries_Dumor_Yao_20200621_1901.csv.gzip",compression = "gzip")
print(f"Number of BRI countries: {len(df_bri_countries)}")
df_bri_countries["BRI_membership"] = 1

try:
    df_bri_countries.drop(["Unnamed: 0","BRI_Country"], axis =1, inplace= True)
except:
    pass

df_bri_countries.head()
```

<a id="ID_part3"></a>
### Part 3 | Merge and prep data
|| [0| Default imports](#ID_top) || [1|Part1](#ID_part1) || [2|Part2](#ID_part2) || [3|Part3](#ID_part3) || [4|Part4](#ID_part4) || [5|Part5](#ID_part5) ||

```python
gdf_map_merged = gdf_map.merge(df_bri_countries, on = "iso_3",how = "left",copy = False)
#gdf_map_merged["BRI_membership"].fillna(0,inplace= True)
print(gdf_map_merged.BRI_membership.value_counts())
#gdf_map_merged.BRI_membership.replace(0,2,inplace= True)

gdf_map_merged["BRI_label"] = gdf_map_merged["BRI_membership"].copy()
#gdf_map_merged.BRI_label.replace(0,"Non-participating",inplace= True)
gdf_map_merged.BRI_label.replace(1,"2020 Participating",inplace= True)
```

```python
# Colour column
colour_series = []

for entry in list(gdf_map_merged.BRI_membership):
    
    if entry == 1:
        colour_series.append("#aa381e")
    else:
        colour_series.append("white")

gdf_map_merged["colour"] = colour_series
gdf_map_merged.head()
```

```python
# sample plot
gdf_map_merged.plot()
```

```python
# dataframe to feed plot
gdf_plot = gdf_map_merged[(gdf_map_merged.country_name!="Antarctica")]
number_categories = len(gdf_plot.BRI_label.unique())

# Plotting # define number of bins we want from cmap
fig, ax_bri = plt.subplots(1, figsize=(25,20))

gdf_plot.plot(
    # Data
    column = "BRI_label",categorical = True,
    
    # Colouring map and data
    cmap= plt.cm.get_cmap("Pastel2",number_categories),
    #color = gdf_plot.colour,
    linewidth = 0.2,
    edgecolor = "darkgreen",
    hatch = "",
    
    # Missing data
    missing_kwds={
            "color": "lightgrey",
            #"edgecolor": "lightcoral",
            "edgecolor": "dimgrey",
            "linewidth": 0.2,
            #"hatch": "////",
            "label": "Missing values",
    },
    # Legend
    legend= True,legend_kwds={'bbox_to_anchor':(0.2,0.2),'fontsize':15,'frameon':False},
    
    # Assign to axis
    ax =ax_bri)

# Labels
gdf_plot.apply(
    lambda x: ax_bri.annotate(s=x.iso_3, xy=x.geometry.centroid.coords[0], ha='center',
                              color = "black",fontsize = 6),axis=1)

# Final customisation (map axis, title and layout padding)
ax_bri.axis('on')
ax_bri.set_title("Belt and Road Initiative participation",fontsize=30)
plt.tight_layout()
```

```python
# Save image
keyword = "_no_hatch"
fig.savefig(f"../Images/BRI_members{keyword}.png",dpi=300)
```
